```python
# ======================================== IMPORTING REQUIRED PYTHON MODULES ===================================== #
#Importing Pandas module
import pandas as pd
#Importing train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split
#Importing LinearRegression class from the sklearn.linear_model
from sklearn.linear_model import LinearRegression
#To evalute the regression model using R2 squared
from sklearn.metrics import r2_score
#Import pickle module to store the model in file
import pickle

# ======================================= MAIN CODE ========================================================= #
#Using pandas module reading the dataset from the input csv file
dataset = pd.read_csv("Salary_Data.csv")
#print the dataset which is stored in dataset variable
print("Whole dataset: \n", dataset)

# Fetching Independent data from the dataset
independent = dataset[["YearsExperience"]]
# Fetching Dependent data from the dataset
dependent = dataset[["Salary"]]

# =================== SPLIT THE TRAIN AND TEST SET BASED ON THE RATIO PROVIDED =========================== #
#Splitting the train set as 70% and test set as 30% with fixed random seed 
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)

# ================== CREATING LINEAR MODULE ======================================= #
#Creating object for LinearRegression
regressor = LinearRegression()
#Create a model fit function by passing X_train (independent training set) and y_train (dependent training set)
regressor.fit(X_train, y_train)

# ================== FROM CREATED MODEL, FIND THE SLOPE AND BIAS/INTERCEPT VALUE ================== #
#Find the slope value
weight = regressor.coef_
#Find the bias/intercept value
bias = regressor.intercept_

# ================= FIND THE PREDICTED VALUE FOR NEW INPUT DATA FOR TRAINING ================ #
#Using trained model (regressor) which is created above, store the predicted values for X_test list.
y_prediction = regressor.predict(X_test)

# =========================== CALCULATE R2 SQUARE SCORE ============================ #
#Calculate the R2 score using target y_test and prediction y_prediction data
r_score = r2_score(y_test, y_prediction)
#print the r_score value
print(r_score)

# ========================== SAVE THE MODEL =================================== #
#save the model in file with extention .sav
filename = "finalized_model_linear.sav"
pickle.dump(regressor, open(filename, 'wb'))

# ========================= DEPLOY MODEL ============================ #
loaded_data = pickle.load(open(filename,'rb'))

# ======================== EVALUATE THE LOADED MODULE BY INPUT =========================== #
result = loaded_data.predict([[15]])
print("Predicted Output value: ", result)
```
