```python
import pandas as pd
```

```python
input_dataset = pd.read_csv("insurance_pre.csv")
input_dataset
```

```python
input_dataset = pd.get_dummies(input_dataset, dtype=int, drop_first=True)
input_dataset
```

```python
input_dataset.columns
```

```python
independent = input_dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
dependent = input_dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, y_train)
```

```python
slope = regressor.coef_
slope
```

```python
bias = regressor.intercept_
bias
```

```python
y_prediction = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_square_score = r2_score(y_test,y_prediction)
r_square_score
```

```python
import pickle
filename = 'finalized_model_MLR.sav'
pickle.dump(regressor,open(filename, 'wb'))
```
