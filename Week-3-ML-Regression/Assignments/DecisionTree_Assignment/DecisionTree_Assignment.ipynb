```python
import pandas as pd
```

```python
input_dataset = pd.read_csv("insurance_pre.csv")
input_dataset
```

```python
input_dataset = pd.get_dummies(input_dataset, dtype=int, drop_first=True)
input_dataset
```

```python
input_dataset.columns
```

```python
independent = input_dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = input_dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='absolute_error',max_features='sqrt',splitter='best')
regressor = regressor.fit(X_train, y_train)
```

```python
#import matplotlib.pyplot as plt
#from sklearn import tree
#tree.plot_tree(regressor)
#plt.show()
```

```python
y_prediction = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test, y_prediction)
r_score
```

```python
import pickle
filename = 'finalized_model_DT.sav'
pickle.dump(regressor,open(filename,'wb'))
```
