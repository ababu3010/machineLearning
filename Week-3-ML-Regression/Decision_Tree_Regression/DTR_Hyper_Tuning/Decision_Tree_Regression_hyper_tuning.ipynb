```python
import pandas as pd
```

```python
input_data = pd.read_csv("50_Startups.csv")
input_data
```

```python
input_data = pd.get_dummies(input_data, dtype=int, drop_first=True)
input_data
```

```python
input_data.columns
```

```python
independent = input_data[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
dependent = input_data[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='squared_error', max_features='log2', splitter='random')
regressor = regressor.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_predict = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_predict)
```
